# Project Description

1. Pixel values of images normalized by dividing by 255, which scales pixel values to range of [0, 1].
2. List of class names defined, representing different categories in CIFAR-10 dataset.
3. A loop is executed to display a 4x4 grid of sample images from the training set.
4. The displayed plot of sample images is shown using plt.show().
5. Training and testing sets are trimmed down to reduce the number of samples. 
6. A pre-trained model is loaded using the models.load_model() function.
7. An image is read using OpenCV's cv.imread() function
8. Color channels of image converted from BGR to RGB using OpenCV's cv.cvtColor() function.
9. The RGB image is displayed using plt.imshow().
10. Image resized to dimensions expected by model (32x32) using OpenCV's cv.resize() function.
11. Image is normalized by dividing by 255.
12. Model predicts class probabilities for resized image using model.predict(). Resized image is wrapped in NumPy array and divided by 255 to ensure has same normalization as training and testing sets.
13. Index of class with highest probability is determined using np.argmax().

```python
#Import Libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras import datasets, layers, models
```

```python
(training_images, training_labels), (testing_images, testing_labels) = datasets.cifar10.load_data()
```

```python
training_images, testing_images = training_images / 255, testing_images / 255
```

```python
class_names = ['Plane', 'Car', 'Bird', 'Cat', 'Deer', 'Dog', 'Frog', 'Horse', 'Ship', 'Truck']
```

```python
for i in range(16):
    plt.subplot(4, 4, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(training_images[i], cmap = plt.cm.binary)
    plt.xlabel(class_names[training_labels[i][0]])

plt.show()
```

```python
training_images = training_images[:20000]
training_labels = training_labels[:20000]
testing_images = testing_images[:4000]
testing_labels = testing_labels[:4000]
```

```python
model = models.load_model(r'C:\Users\jaxz2\OneDrive\AI&ML\Projects\imageclassification\image_classifier.model')
```

```python
img = cv.imread(r'C:\Users\jaxz2\OneDrive\AI&ML\Projects\imageclassification\1.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img, cmap = plt.cm.binary)
```

```python
resized_img = cv.resize(img, (32, 32))
prediction = model.predict(np.array([resized_img]) / 255)
index = np.argmax(prediction)
print(f"Prediction is {class_names[index]}")
```
